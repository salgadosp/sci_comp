```python
def primes(n):
    if n < 1:
        return []

    primes = []
    limit = 15  # initial estimate (we'll adjust if we need more numbers)
    
    while len(primes) < n:
        sieve = [True] * (limit + 1)
        sieve[0], sieve[1] = False, False  # 0 and 1 are not prime
        
        for i in range(2, int(limit**0.5) + 1):
            if sieve[i]:
                for multiple in range(i * i, limit + 1, i):
                    sieve[multiple] = False
        
        primes = [num for num, is_prime in enumerate(sieve) if is_prime]
        limit *= 2  # double the limit if we haven't found enough primes

    return primes[:n]
```

```python
prime_list = primes(1000000)
```

```python
def first_n_ks(n):
    # Generate the first n primes
    primes_list = primes(n)
    
    # Calculate each k value for primes 1 through n
    k_values = [primes_list[i] / ((i + 1) * log(i + 1)) for i in range(1, n)]
    
    return k_values

# Example usage
n = 100  # Specify how many k values you want
k_values = first_n_ks(n)
```
